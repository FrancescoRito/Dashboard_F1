```python
import json as j
import pandas as pd
import requests as r


#Get Connection with the API
base_url = "http://ergast.com/api/f1/circuits.json"

URL0= r.get(base_url)

info=URL0.json()

#Json the api link to get all the data
info=URL0.json()

total=info['MRData']['total']

#Get a new api connection with the total of the data
URL1="http://ergast.com/api/f1/circuits.json?limit="+total

URL=r.get(URL1)

#Import of the data in json format
circuits_info=URL.json()

#Create lists for the final dataframe
circuitId = []
circuitName = []
locality = []
country = []
latlong = []

#Looping the json file to get all the data
for circuit in circuits_info['MRData']['CircuitTable']['Circuits']:
   circuitId.append(circuit['circuitId'])
   circuitName.append(circuit['circuitName'])
   locality.append(circuit['Location']['locality'])
   country.append(circuit['Location']['country'])
   latlong.append(circuit['Location']['lat']+','+circuit['Location']['long'])

#Create the dataframe
circuits_df = pd.DataFrame({ 'circuitId' : circuitId,
                             'circuitName' : circuitName,
                             'locality' : locality,
                            'country' : country,
                            'lat,long' : latlong
                            })
circuits_df
```
